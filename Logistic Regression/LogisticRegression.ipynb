**Logistical Regression**

This ML algorithm helps us in classification and provides us binary lables as answer.

More or less the dw and db update method is the same, the only difference is as follows:

    1. We introduce the sigmoid function:

        sigmoid(z) = 1/(1+ np.exp(-z)) where,
        
        z = np.dot(X, weights) + bias
    
    2. The sigmoid function helps us keeping the value of z between 0 and 1, hence helping us to put label based upon a certain threshold(0.5 in our case).

```python
import numpy as np
import math
```

```python
class Vector:
    def __init__(self, data):
        self.data = data
        self.dim = len(data)

    def __repr__(self):
        return f"The vector is {self.data}"
    def __len__(self):
        return self.dim

    def dot(self, other): #We use the "other" argument--> to let the function know that its an object of same class.
        dot = 0
        if self.dim != other.dim:
            raise ValueError(f"Dimensions of the vectors are not the same: {self.dim} vs {other.dim}")
        else:
            for i in range(self.dim):
                dot += self.data[i]*other.data[i]
            return dot

    def norm(self):
        return mt.sqrt(Vector.dot(self,self))

    def add(self,other):
        if self.dim != other.dim:
            raise ValueError(f"The dimensions of the vectors are not the same: {self.dim} vs {other.dim}")
        else:
            a = []
            for i in range(self.dim):
                a.append([[]])
                a[i] = self.data[i] + other.data[i]
        return Vector(a)
    def scale(self,scaler):
        a = []
        for i in range(self.dim):
            a.append([[]])
            a[i] = scaler*self.data[i]
        return Vector(a)
```

```python
class LogisticRegression():
  def __init__(self, lr=0.01, epochs = 1000):
    self.lr = lr
    self.epochs = epochs
    self.bias = None
    self.weights = None
  def sigmoid(self, z):
    fun = 1/(1+np.exp(-z))
    return fun
  def fit(self, X, y):
    n_samples, n_features = X.shape
    self.bias = 0
    self.weights = np.zeros(n_features)
    for i in range(self.epochs):
      z = np.dot(X, self.weights) + self.bias
      y_pred= self.sigmoid(z)
      dw = (1/n_samples)*np.dot(X.transpose(), (y_pred - y))
      db = (1/n_samples)*np.sum(y_pred - y)
      self.bias -= self.lr*db
      self.weights -= self.lr*dw
  def predict(self,X):
    z = np.dot(X, self.weights) + self.bias
    f = self.sigmoid(z)
    return (f>=0.5).astype(int) #Returns the values for all the elements in the array
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, random_state=42)

model = LogisticRegression(lr=0.1, epochs=1000)
model.fit(X, y)

y_pred = model.predict(X)

from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y, y_pred))
```
