**Whats the difference between random forests and decision tree?**

Chances of biasness in decision trees is quite high cause we usually train them on the complete training set, so in order to lower the gini, they tend to overfit/ memorize the training set. In order to solve this issue, we try to randomly assign the dataset using the following function(which also may have restrictions on the length of the instances to be taken) with replacement, in order to create random sets of training data for each tree, hence reducing the baisness, and as it includes many such trees based upon randomness, we call it **Random Forests**.

```python
class RandomForest:
    def __init__(self, n_trees=10, max_depth=10, min_samples_split=2, n_features=None):
        self.n_trees = n_trees
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.n_features = n_features
        self.trees = []

    def fit(self, X, y):
        for i in range(self.n_trees):
            X_sample, y_sample = self.bootstrap(X, y)
            tree = DecisionTree(self.max_depth)
            tree.fit(X_sample, y_sample)
            self.trees.append(tree)

    def bootstrap(self, X, y):
        indices = np.random.choice(len(X), size=len(X), replace= True)
        return X[indices], y[indices] #Returns the indices that have been randomly choosen.

    def predict(self, X):
        predictions = [tree.predict(X) for tree in self.trees]
        predictions = np.array(predictions)  # shape: (n_trees, n_samples)
        # for each sample, find most common label across all trees
        return np.array([self._most_common_label(predictions[:, i]) for i in range(X.shape[0])])

    def _most_common_label(self, y):
        values, counts = np.unique(y, return_counts=True)
        return values[np.argmax(counts)]
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=300, noise=0.2, random_state=42)

rf = RandomForest(n_trees=10, max_depth=5)
rf.fit(X, y)
preds = rf.predict(X)
print(f"Accuracy: {np.mean(preds == y) * 100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Generate data
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                            random_state=42, n_clusters_per_class=1)

# Train your RF
clf = RandomForest(n_trees=10, max_depth=5)
clf.fit(X, y)

# Create mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict on mesh
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.4, cmap='RdYlBu')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='RdYlBu', edgecolors='k')
plt.title("Random Forest Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
