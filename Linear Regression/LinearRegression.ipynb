```python
import math as mt
import numpy as np
```

```python
class Vector:
    def __init__(self, data):
        self.data = data
        self.dim = len(data)

    def __repr__(self):
        return f"The vector is {self.data}"
    def __len__(self):
        return self.dim

    def dot(self, other): #We use the "other" argument--> to let the function know that its an object of same class.
        dot = 0
        if self.dim != other.dim:
            raise ValueError(f"Dimensions of the vectors are not the same: {self.dim} vs {other.dim}")
        else:
            for i in range(self.dim):
                dot += self.data[i]*other.data[i]
            return dot

    def norm(self):
        return mt.sqrt(Vector.dot(self,self))

    def add(self,other):
        if self.dim != other.dim:
            raise ValueError(f"The dimensions of the vectors are not the same: {self.dim} vs {other.dim}")
        else:
            a = []
            for i in range(self.dim):
                a.append([[]])
                a[i] = self.data[i] + other.data[i]
        return Vector(a)
    def scale(self,scaler):
        a = []
        for i in range(self.dim):
            a.append([[]])
            a[i] = scaler*self.data[i]
        return Vector(a)
```

```python
class Matrix:
    def __init__(self, data):
        self.data = data
        self.rows = len(data)
        self.columns = len(data[0])
        for row in self.data:
            if len(row) != self.columns:
                raise ValueError(f"All rows must have the same number of Columns!!!")
    def __repr__(self):         #Must return a string
        result = ""
        for row in self.data:
            result += f"{row}\n"
        return result

    def transpose(self):
        a = []
        for i in range(self.columns):
            row = []
            for j in range(self.rows):
                row.append(self.data[j][i])
            a.append(row)
        return Matrix(a)

    def multiply(self, other):
        if self.columns != other.rows:
            raise ValueError(f"The number of rows: {other.rows} and the number of columns: {self.columns} dont match!!!")
        else:
            result = []
            for i in range(self.rows):
                row = []
                for j in range(other.columns):
                    row_i = Vector(self.data[i])
                    col_j = Vector([other.data[k][j] for k in range(other.rows)])
                    row.append(row_i.dot(col_j))
                result.append(row)
            return result
```

```python
class LinearRegression():
  def __init__(self,lr=0.01,epochs=1000):
    self.lr = lr
    self.epochs = epochs
    self.weights = None
    self.bias = None
  def fit(self,X,y):
    n_samples, n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0
    for i in range(self.epochs):
      y_pred = np.dot(X,self.weights) + self.bias
      dw = (1/n_samples)*np.dot(X.transpose(),(y_pred - y))
      db =  (1/n_samples)*np.sum(y_pred - y)
      self.weights -=self.lr*dw
      self.bias -=self.lr*db
    return np.dot(X,self.weights) + self.bias
  def predict(self, X):
    y_predict = np.dot(X, self.weights) + self.bias
    return y_predict
```

**Synthetic Data**

```python
# Synthetic data — true relationship: y = 3x + 4
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X.squeeze() + np.random.randn(100)

# Train
model = LinearRegression(lr=0.01, epochs=1000)
model.fit(X, y)

print("Weight:", model.weights)  # should be close to 3
print("Bias:  ", model.bias)     # should be close to 4
```
